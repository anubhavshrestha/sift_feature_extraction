```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def show_sift_image(image_path):
    # Check if the file extension is either .jpg or .png
    if image_path.endswith('.jpg') or image_path.endswith('.png') or image_path.endswith('.jpeg'):
        # Load the image
        image_np = cv2.imread(image_path)

        if image_np is None:
            print("Error: Unable to load image.")
            return
    else:
        print("Error: Input file must be a .jpg or .png file.")
        return

    # Convert image to grayscale
    gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

    # Initialize SIFT detector
    sift = cv2.SIFT_create()

    # Detect keypoints
    kp = sift.detect(gray, None)

    # Draw keypoints on the image
    sift_image = cv2.drawKeypoints(image_np, kp, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    # Display the image with SIFT keypoints
    plt.imshow(cv2.cvtColor(sift_image, cv2.COLOR_BGR2RGB))
    plt.title("SIFT Keypoints")
    plt.axis('off')
    plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def show_image(image_path):
    try:
        image = Image.open(image_path)
        plt.imshow(image)
        plt.axis('off')  # Hide axis
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

# Example usage:
# Provide the path to the image file
show_image("/content/dog.jpeg")
```

```python
import cv2
import matplotlib.pyplot as plt

def detect_and_draw_keypoints(image_path):
    # Load the image
    img = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Initialize SIFT detector
    sift = cv2.SIFT_create()

    # Detect keypoints
    kp = sift.detect(gray, None)

    # Draw keypoints on the image
    img_with_keypoints = cv2.drawKeypoints(gray, kp, img, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    # Convert the image from BGR to RGB
    img_with_keypoints_rgb = cv2.cvtColor(img_with_keypoints, cv2.COLOR_BGR2RGB)

    # Plot the image with keypoints
    plt.imshow(img_with_keypoints_rgb)
    plt.axis('off')
    plt.title('Image with SIFT Keypoints')
    plt.show()

# Example usage:
# Provide the path to the image file
detect_and_draw_keypoints('/content/dog.jpeg')
```
